```python
import pandas as pd
dataset = pd.read_csv('Employee_Salary_Dataset.csv')
```

```python
print('rows:', dataset.shape[0])
print('columns:', dataset.shape[1])
```

```python
dataset.isnull().sum()
```

```python
dataset_filled = dataset.copy()
for col in dataset_filled.columns:
    if dataset_filled[col].dtype in ['int64','float64']:
        dataset_filled[col] = dataset_filled[col].fillna(dataset_filled[col].mean())
    else:
        dataset_filled[col] = dataset_filled[col].fillna(dataset_filled[col].mode()[0])
dataset_filled.head()
```
