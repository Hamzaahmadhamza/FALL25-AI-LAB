```python
import math

def minimax_algorithm(depth, index, is_max, values, height):
    if depth == height:
        return values[index]

    if is_max:
        return max(
            minimax_algorithm(depth + 1, index * 2, False, values, height),
            minimax_algorithm(depth + 1, index * 2 + 1, False, values, height)
        )
    else:
        return min(
            minimax_algorithm(depth + 1, index * 2, True, values, height),
            minimax_algorithm(depth + 1, index * 2 + 1, True, values, height)
        )

values = [3, 5, 2, 9, 3, 5, 2, 9]
tree_height = int(math.log(len(values), 2))
optimal_value = minimax_algorithm(0, 0, True, values, tree_height)
print("Optimal Value:", optimal_value)
```
